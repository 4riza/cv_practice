```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/cat_cats_eyes_curious.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()


a = 1
b = 50.0

new_img = a * img + b

new_img = np.clip(new_img, 0, 255)
new_img = new_img.astype(np.uint8)

plt.imshow(new_img)
plt.title("Adjusted Image")
plt.axis('off')
plt.show()

```
