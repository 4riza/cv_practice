```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read first image
img1 = cv2.imread('/content/cat_cats_eyes_curious.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.imshow(img1)
plt.title("Cat Original")
plt.axis('off')
plt.show()

# Read second image
img2 = cv2.imread('/content/Dog-1123016_960_720.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img2)
plt.title("Dog Original")
plt.axis('off')
plt.show()

# Resize img2 to match img1
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Linear blending
a = 0.5
blended_img = (1 - a) * img1 + a * img2_resized

# Clip and convert
new_img = np.clip(blended_img, 0, 255).astype(np.uint8)

plt.imshow(new_img)
plt.title("Blended Image")
plt.axis('off')
plt.show()
```
